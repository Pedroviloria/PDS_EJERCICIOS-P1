# **Señal aperiodica y desplazada**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir una señal discreta no periódica
n = np.arange(0, 10)
x = np.random.randn(len(n))

# Desplazar la señal
shift = 2
x_shifted = np.zeros_like(x)
x_shifted[shift:] = x[:-shift]

# Graficar ambas señales
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

ax1.stem(n, x, linefmt='C0-', markerfmt='C0o', basefmt='k-')
ax1.set_title('Señal original')
ax1.set_xlabel('n')
ax1.set_ylabel('Amplitud')
ax1.grid(True)

ax2.stem(n, x_shifted, linefmt='C1-', markerfmt='C1o', basefmt='k-')
ax2.set_title(f'Señal desplazada en x(n-{shift}) ' )
ax2.set_xlabel('n')
ax2.set_ylabel('Amplitud')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

# **Señal periodica y desplazada**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir una señal discreta periódica
n = np.arange(0, 10)
x = np.abs(n - 5)

# Desplazar la señal
shift = 3
x_shifted = np.zeros_like(x)
x_shifted[shift:] = x[:-shift]

# Graficar ambas señales
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

ax1.stem(n, x, linefmt='C0-', markerfmt='C0o', basefmt='k-')
ax1.set_title('Señal original')
ax1.set_xlabel('n')
ax1.set_ylabel('Amplitud')
ax1.grid(True)

ax2.stem(n, x_shifted, linefmt='C1-', markerfmt='C1o', basefmt='k-')
ax2.set_title(f'Señal desplazada en x(n-{shift}) ' )
ax2.set_xlabel('n')
ax2.set_ylabel('Amplitud')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

# **Filtro de media movil**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una señal con ruido
n = np.arange(0, 10)
x = np.sin(2*np.pi*5*n/10) + np.random.normal(0, 0.5, 10)

# Definir la longitud de la ventana del filtro
window_len = 5

# Definir el filtro de media móvil sin ventana
def moving_average(x, window_len):
    cumsum = np.cumsum(np.insert(x, 0, 0))
    return (cumsum[window_len:] - cumsum[:-window_len]) / window_len

# Aplicar el filtro de media móvil sin ventana
x_filtered = moving_average(x, window_len)

# Graficar la señal original y la señal filtrada
plt.stem(n, x, linefmt='C0-', markerfmt='C0o', basefmt='k-')
plt.title('Señal original')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

plt.stem(n[window_len-1:], x_filtered, linefmt='C1-', markerfmt='C1o', basefmt='k-')
plt.title('Señal filtrada')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```
