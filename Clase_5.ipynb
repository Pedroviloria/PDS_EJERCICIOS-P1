# **Linealidad e invarianza en el tiempo**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar dos señales discretas
n = np.arange(0, 10)
x1 = np.sin(np.pi/2 * n)
x2 = np.cos(np.pi/4 * n)

# Sumar las señales
y1 = x1 + x2

# Realizar la prueba de linealidad
a = 2
y2 = a * x1 + a * x2

# Realizar la prueba de invarianza en el tiempo
m = 2
y3 = x1 + x2
y4 = x1[m:] + x2[m:]

# Graficar las señales
fig, axs = plt.subplots(3, 2, figsize=(8, 10))

# Señal 1
axs[0, 0].stem(n, x1)
axs[0, 0].set_title('Señal 1')
axs[0, 0].set_xlabel('n')
axs[0, 0].set_ylabel('Amplitud')

# Señal 2
axs[0, 1].stem(n, x2)
axs[0, 1].set_title('Señal 2')
axs[0, 1].set_xlabel('n')
axs[0, 1].set_ylabel('Amplitud')

# Señal sumada
axs[1, 0].stem(n, y1)
axs[1, 0].set_title('Señal sumada')
axs[1, 0].set_xlabel('n')
axs[1, 0].set_ylabel('Amplitud')

# Prueba de linealidad
axs[1, 1].stem(n, y1, label='y1')
axs[1, 1].stem(n, y2, label='y2')
axs[1, 1].set_title('Prueba de linealidad')
axs[1, 1].set_xlabel('n')
axs[1, 1].set_ylabel('Amplitud')
axs[1, 1].legend()

# Prueba de invarianza en el tiempo
axs[2, 0].stem(n, y3, label='y3')
axs[2, 0].set_title('Prueba de invarianza en el tiempo')
axs[2, 0].set_xlabel('n')
axs[2, 0].set_ylabel('Amplitud')

axs[2, 1].stem(n[m:], y4, label='y4')
axs[2, 1].set_xlabel('n')
axs[2, 1].set_ylabel('Amplitud')
axs[2, 1].legend()

plt.tight_layout()
plt.show()
```

# **Convolución**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar señales discretas
n = np.arange(0, 10)
x1 = np.sin(np.pi/2 * n)
x2 = np.cos(np.pi/4 * n)

# Realizar la convolución
y = np.convolve(x1, x2)

# Graficar las señales y la convolución
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Señal 1
axs[0].stem(n, x1)
axs[0].set_title('Señal 1')
axs[0].set_xlabel('n')
axs[0].set_ylabel('Amplitud')
axs[0].grid(True)

# Señal 2
axs[1].stem(n, x2)
axs[1].set_title('Señal 2')
axs[1].set_xlabel('n')
axs[1].set_ylabel('Amplitud')
axs[1].grid(True)

# Convolución
axs[2].stem(np.arange(0, len(y)), y)
axs[2].set_title('Convolución')
axs[2].set_xlabel('n')
axs[2].set_ylabel('Amplitud')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
